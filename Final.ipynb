### Metric
Our metric to decide the best neighborhood was the one that would be most likely to **be saved by a superhero.**

## Analysis

### Arrests in Pittsburgh

This is one of the datasets that I analyzed as a factor to consider where a superhero may more likely to be involved in are: areas where there may be a higher chance of catching someone who may be performing an action to get arrested.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('ArrestsinPittsburgh.csv')
neighborhood_counts = data['INCIDENTNEIGHBORHOOD'].value_counts().reset_index()
neighborhood_counts.columns = ['Neighborhood', 'Arrest Count']
plt.figure(figsize=(20, 10))
plt.bar(neighborhood_counts['Neighborhood'], neighborhood_counts['Arrest Count'])
plt.title('Arrests by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Arrest Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### Fire Incidents in Pittsburgh

This is another one of the datasets that I analyzed as a factor to consider where a superhero may more likely to be involved in an event where they may have to save someone who is caught in a building that is on fire. 

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('fireincidents.csv')

neighborhood_counts = data['neighborhood'].value_counts()

plt.figure(figsize=(20, 10))
neighborhood_counts.plot(kind='bar')
plt.title('Fire Incidents by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Fire Count')
plt.tight_layout()
plt.show()
```

### Police Incidents in Pittsburgh

This is one of the datasets that I analyzed as a factor to consider where a superhero may more likely to be involved in are areas where there may be more incidents with police as they could have a higher chance of being able to help the police if they are in the area.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('PoliceIncidentData.csv')

neighborhood_counts = data['INCIDENTNEIGHBORHOOD'].value_counts()

plt.figure(figsize=(20, 10))
neighborhood_counts.plot(kind='bar')
plt.title('Police Incidents by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Police Incident Count')
plt.tight_layout()
plt.show()
```

### Final Analysis

In the final analysis, I brought together the graphs and limited the amount of neighborhoods shown to 20 for easier readibility and visibility of the neighborhood most likely to be saved by a superhero. Our metric for figuring this out was by simplying displaying the graphs above on top of each other to then determine which one would be on the highest compared to the rest.

```python
import pandas as pd
import matplotlib.pyplot as plt

arrests_data = pd.read_csv('ArrestsinPittsburgh.csv')
fire_data = pd.read_csv('fireincidents.csv')
police_data = pd.read_csv('PoliceIncidentData.csv')

arrests_counts = arrests_data['INCIDENTNEIGHBORHOOD'].value_counts().head(25)
fire_counts = fire_data['neighborhood'].value_counts().head(25)
police_counts = police_data['INCIDENTNEIGHBORHOOD'].value_counts().head(25)

plt.figure(figsize=(20, 10))

arrests_counts.plot(kind='bar', color='blue', position=0, width=0.25, label='Arrests')
fire_counts.plot(kind='bar', color='red', position=1, width=0.25, label='Fire Incidents')
police_counts.plot(kind='bar', color='green', position=2, width=0.25, label='Police Incidents')

plt.title('Incidents by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

### Conclusions

From our findings, the best neighborhood to most likely be saved or patrolled by a superhero is Central Business District. It is the neighborhood with the most counts of police incidents, fire incidents, and arrests in Pittsburgh.
