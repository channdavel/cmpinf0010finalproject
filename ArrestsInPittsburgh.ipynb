I listed a count of each neighborhood that had an arrest in descending order to help determine the best neighborhood.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('ArrestsinPittsburgh.csv')
neighborhood_counts = data['INCIDENTNEIGHBORHOOD'].value_counts().reset_index()
neighborhood_counts.columns = ['Neighborhood', 'Arrest Count']
plt.figure(figsize=(20, 10))
plt.bar(neighborhood_counts['Neighborhood'], neighborhood_counts['Arrest Count'])
plt.title('Arrests by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Arrest Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

From this graph, I can say that the neighbourhood with the most arrests is Central Business District. Then the next four that follow it are: South Side FLats, Carrick, East Allegheny, and Homewood South. This can more or less coincide with my best neighbourhood metric as the superheros are most likely to help neighbourhoods with the most arrests as it can be tied in with that neighbourhood having a high crime rate around it.
