# Import data

```python
import pandas as pd

data = pd.read_csv('fireincidents.csv')
neighborhood_counts = data['neighborhood'].value_counts().reset_index()
neighborhood_counts.columns = ['neighborhood', 'fire_count']
print(neighborhood_counts)
```

I formatted the neighborhoods in descending order by counting how often a neighborhood had a fire occur. This would then print out the formatted data set with the neighborhoods that have the highest and lowest amount of fires occur.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('fireincidents.csv')

neighborhood_counts = data['neighborhood'].value_counts()

plt.figure(figsize=(20, 10))
neighborhood_counts.plot(kind='bar')
plt.title('Fire Incidents by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Fire Count')
plt.tight_layout()
plt.show()
```

This is a graphical representation of all of the neighborhoods in descending order by the count of how many fires have occurred. 

From this graph, I can say that the neighbourhood with the most fires is Carrick and the one with the least fires is Mount Oliver Borough. This is an important factor to consider for our best neighbourhood metric. Since Carrick, Central Business District, South Side Flats, and Squirrel Hill South have a lot of fires, it can help us decide which one a superhero might decide to pop in and help save them. This would mean that more or less the top 10 neighbourhoods with the highest fire incidents are most likely to be saved by a superhero.
